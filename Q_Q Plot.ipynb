```python
# Q_Q Plot

import numpy as np
import pylab
import scipy.stats as stats

# N(0,1)
std_normal=np.random.normal(loc=0,scale=1,size=1000)

# 0 to 100 percentile of standard_normal
for i in range(0,101):

    print(i,np.percentile(std_normal,i))

    
```

```python
# generates 100 samples from N(0,1)

measurement=np.random.normal(loc=20, scale=5,size=100)
stats.probplot(measurement,dist="norm",plot=pylab)
pylab.show()
```
