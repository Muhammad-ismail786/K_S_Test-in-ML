```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

# generate a Gussian r.v x
      
x=stats.norm.rvs(size=1000);
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw_method=0.5)
plt.show()
stats.kstest(x,'norm')
    
```

```python
# y_continus uniform distribution(0,1)

y=np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y),bw_method=0.5)
plt.show()
stats.kstest(y,'norm')
```
