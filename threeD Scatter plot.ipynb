```python
# HISTOGRAM,CDF,PDF

# what about 1_D scatter plot using just one features
# 1_D scatter plot of petal_length


import numpy as np
iris=pd.read_csv("iris.csv")
iris_setosa=iris.loc[iris["species"]=="setosa"];
iris_virginica=iris.loc[iris["species"]=="virginica"];
iris_versicolor=iris.loc[iris["species"]=="versicolor"];

# print (iris_setosa["petal_length"]) 

plt.plot(iris_setosa["petal_length"],np.zeros_like(iris_setosa['petal_length']))
plt.plot(iris_versicolor["petal_length"],np.zeros_like(iris_versicolor['petal_length']))
plt.plot(iris_virginica["petal_length"],np.zeros_like(iris_virginica['petal_length']))
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",height=5)\
   .map(sns.histplot,"petal_length")\
   .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",height=5)\
   .map(sns.histplot,"sepal_length")\
   .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",height=5)\
   .map(sns.histplot,"sepal_width")\
   .add_legend();
plt.show()
```

```python
sns.FacetGrid(iris,hue="species",height=5)\
   .map(sns.histplot,"petal_width")\
   .add_legend();
plt.show()
```

```python
# from the above histogram 
# we can take a results that 
# petal_length technique is much better than other four bcz the species are well separated to each other(setosa,versicolor,virginica)
# on the basis of result
# sl>pw>>(muvh bettet than) sl>>sw
# so on the basis of accuracy we can pick(petal_length)
# after that for two features we can pick petal_length and petal_width
```

```python
# NEED FOR CUMMULATIVE DISTRIBUTION FUNCTION(CDF)
# we can visually see what percentage of versicolor flowers have a
# petal_length of less than 1.5 ?
# How to costruct a CDF
# HOW to read a CDF

# plot CDF of petal_length

counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10, density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

# COMPUTE PDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()

```

```python
# plot of CDF of petal_length of different flowers
# Misclassification error if you use petal_length only

counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#VIRGINICA

counts,bin_edges=np.histogram(iris_virginica['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

# VERSICOLOR
counts,bin_edges=np.histogram(iris_versicolor['petal_length'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()




```

```python
# mean, variance, std_deviation
print("\n Means:")
print(np.mean(iris_setosa['petal_length']))
# mean with an outlier
print(np.mean(np.append(iris_setosa['petal_length'],50)))
print(np.mean(iris_virginica['petal_length']))
print(np.mean(iris_versicolor['petal_length']))

print("\nStd_dev:")
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_virginica['petal_length']))
print(np.std(iris_versicolor['petal_length']))

```

```python
#median :

#First sort items incr/decr order
#pick the middle value
#median will not get corrupted due to the some outliers, but if more than 50 percent values are currupted then the median gets currupted.
print("\n Medians:")
print(np.median(iris_setosa['petal_length']))
print(np.median(np.append(iris_setosa['petal_length'],50))) # median with an outlier
print(np.median(iris_virginica['petal_length']))
print(np.median(iris_versicolor['petal_length']))
```

```python
#Percentile :
#If the value is at 5Oth position in an given array it is called as 50th percentile. It tells us than about 50 percent values are less than this value and about 50 percent values are greater than this value.
#Median = 50th percentile value of x = Xs[50] where Xs is an array
#25th , 50th, 75th and 100th percentiles are called Quantiles
#Amazon Delivery Example : To understand the number of days it takes for the delivery. When we check 95th or 99th percentile of delivery days, it gives us the insights about how many days it takes to deliver a package for 95 or 99 % times.

print("\n Quantiles:")
print(np.percentile(iris_setosa['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_virginica['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25)))
```

```python
print("\n90th Percentiles:")
print(np.percentile(iris_setosa['petal_length'],90))
print(np.percentile(iris_virginica['petal_length'],90))
print(np.percentile(iris_versicolor['petal_length'],90))
```

```python
# Median Absolute Deviation
from statsmodels import robust
print("\nmedian absolute deviation:")
print(robust.mad(iris_setosa['petal_length']))
print(robust.mad(iris_virginica['petal_length']))
print(robust.mad(iris_versicolor['petal_length']))

```

```python
# Box_plot with wiskers: another method of visualization the 1_D scatter plot
# The concept of Median,Quantile and Percentile.
# How to draw the box in the Box_plot
# How to draw wiskers [no standard way] could use min and max or use other
# NOTE: in the plot given below a techniques inter_quartile range is used in
# wiskers in the plot below do not correspond to the min and max values.
# Box_plot can be visualized as a PDF on the side_ways.

sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
# VIOLIN
# A violin plot combines the benefits of the previous two plots and simplifies them
# Danser region of the data are fatter, and sparser one thinner 
# in a violin plot
#NOTE:
#Best of both worlds - mean(PDF) as well as median(Qunatiles)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
iris=pd.read_csv("iris.csv")
sns.violinplot(x='species',y='petal_length', data=iris,height=8)


plt.grid()
plt.show()
```
